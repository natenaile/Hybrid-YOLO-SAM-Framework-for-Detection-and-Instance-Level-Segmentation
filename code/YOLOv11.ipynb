```python
# yolo11n
from ultralytics import YOLO

model = YOLO("yolo11n.pt")
```

```python
# Train Data
results = model.train(
    data=r"C:\Users\aaa\Documents\DATASETS\EgoHands Public.v1-specific.yolov11\data.yaml",
    epochs=50,
    batch=16,
    imgsz=640,
    workers=4
)
```
