```python
from ultralytics import RTDETR

model = RTDETR("rtdetr-l.pt")
```

```python
##Training Data
results = model.train(
    data="/content/drive/MyDrive/EgoHands Public.v1-specific.rtdetr-l/data.yaml",
    epochs=30,
    batch=16,
    imgsz=640,
    project="/content/drive/MyDrive/result1"
)
```

```python
#Test Data
from ultralytics import RTDETR

model = RTDETR("/content/drive/MyDrive/result1/train/weights/best.pt")
test_images_path = "/content/drive/MyDrive/EgoHands Public.v1-specific.rtdetr-l/test/images"

results = model.predict(
    source=test_images_path,
    save=True,
    save_txt=True,
    save_conf=True,
    project="/content/drive/MyDrive/result1",
    name="test_results"
)

print("Testing completed for the entire test dataset!")
```

```python
#Test Data
model = RTDETR("/content/drive/MyDrive/result1/train/weights/best.pt")

metrics = model.val(
    data="/content/drive/MyDrive/EgoHands Public.v1-specific.rtdetr-l/data.yaml",
    split="test",
    project="/content/drive/MyDrive/result1",
    name="val1"
)
print("Model Evaluation Metrics on Test Data:", metrics)
```
